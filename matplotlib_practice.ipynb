```python
import matplotlib
%matplotlib inline
```

```python
#Import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
#Set x-axis to a list of strings for each month.
x_axis= ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sept","Oct","Nov","Dec"]
```

```python
#Set y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.9,20.71,20.09]
```

```python
#Create a plot
plt.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
#plt.subplots() function returns a tuple that contains a figure and axes objects.
#fig, ax = plt.subplots() unpacks the tuple into variables fing and ax.
#fig references the figure.
#ax references the axes of the figure and allows us to annotate the figure.
#ax.plot() plots the data from x and y axes.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#create the plot
#plt.show() should only be used once per session because it can cause the graph to display improperly.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2,label='Boston')
#create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the y limit betwn 0 and 45.
plt.ylim(0,45)
#Create a title
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Creating a bar chart.
plt.bar(x_axis, y_axis, color="green", label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Creating a horizontal bar chart.
plt.barh(y_axis,x_axis , color="green", label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
#this sentence invert the y axis.
plt.gca().invert_yaxis()
```

```python
#Create a scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
#Create a scatter plot
plt.scatter(y_axis, x_axis, c='red', label='Chicago')
plt.legend()
plt.title('Fare per Month')
plt.ylabel('Months')
plt.xlabel('Fare($)')
plt.gca().invert_yaxis()
plt.xlim(0,45)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Declare an empty list to storage the values 3 times "y" values.
#iterate thru the y_axis to multiply by 3 its values.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
#Using the list comprehension method
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
#Create scatter plot using the Object-Oriented Interface.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#Create bubble chart using Oriented-Object interface.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Creating pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#Wide the entire graph and exlpode the July value.
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#Create the same Pie Chart Using the Object-Oriented Interface.
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```
